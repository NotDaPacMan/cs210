```python
df_hourly = client.get_dataset(
    dataset="caiso_lmp_real_time_15_min",
    start="Jan 1, 2022",
    end="Jan 1, 2023",
    resample="1 hour",
    timezone="US/Pacific",
    filter_column="location",
    filter_operator="in",
    filter_value=["TH_NP15_GEN-APND", "TH_SP15_GEN-APND", "TH_ZP26_GEN-APND"],
)
df_hourly.head(10)
```
